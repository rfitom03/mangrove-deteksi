```python
def list_images_and_labels(root_dir: str) -> Tuple[List[str], List[str]]:
    """Scan subfolder sebagai label; kembalikan list filepath dan label_name."""
    exts = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}
    filepaths, labels = [], []
    if not os.path.isdir(root_dir):
        raise RuntimeError(f"Folder dataset tidak ditemukan: {root_dir}")
    for label_name in sorted(os.listdir(root_dir)):
        class_dir = os.path.join(root_dir, label_name)
        if not os.path.isdir(class_dir):
            continue
        for fname in os.listdir(class_dir):
            ext = os.path.splitext(fname)[1].lower()
            if ext in exts:
                filepaths.append(os.path.join(class_dir, fname))
                labels.append(label_name)
    if len(filepaths) == 0:
        raise RuntimeError(f"Tidak ada gambar di {root_dir}. Pastikan struktur folder per kelas sudah benar.")
    return filepaths, labels


def segment_leaf_mask(bgr: np.ndarray) -> np.ndarray:
    """Membuat mask daun dari gambar BGR (0/255)."""
    if bgr is None or bgr.size == 0:
        return None
    h, w = bgr.shape[:2]
    hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)
    S = hsv[:, :, 1]
    V = hsv[:, :, 2]

    _, th_s = cv2.threshold(S, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    _, th_v = cv2.threshold(V, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    mask = cv2.bitwise_or(th_s, th_v)

    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=2)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=1)

    cnts, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(cnts) == 0:
        return mask
    areas = [cv2.contourArea(c) for c in cnts]
    c = cnts[int(np.argmax(areas))]

    out_mask = np.zeros((h, w), dtype=np.uint8)
    cv2.drawContours(out_mask, [c], -1, 255, thickness=cv2.FILLED)

    if cv2.contourArea(c) < MIN_AREA_RATIO * (h * w):
        out_mask = mask
    return out_mask


def shape_features(mask: np.ndarray) -> Dict[str, float]:
    """Fitur bentuk: area, perimeter, form factor, aspect ratio, extent, solidity, eccentricity, Hu(7)."""
    h, w = mask.shape[:2]
    cnts, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(cnts) == 0:
        return {k: 0.0 for k in [
            "area", "perimeter", "form_factor", "aspect_ratio", "extent",
            "solidity", "eccentricity", *[f"hu_{i+1}" for i in range(7)]
        ]}
    areas = [cv2.contourArea(c) for c in cnts]
    c = cnts[int(np.argmax(areas))]

    area = cv2.contourArea(c)
    perimeter = cv2.arcLength(c, True)
    form_factor = 4 * math.pi * area / (perimeter ** 2 + 1e-8)

    x, y, bw, bh = cv2.boundingRect(c)
    aspect_ratio = bw / (bh + 1e-8)
    extent = area / (bw * bh + 1e-8)

    hull = cv2.convexHull(c)
    hull_area = cv2.contourArea(hull)
    solidity = area / (hull_area + 1e-8) if hull_area > 0 else 0.0

    if len(c) >= 5:
        (xc, yc), (MA, ma), angle = cv2.fitEllipse(c)
        a = max(MA, ma) / 2.0
        b = min(MA, ma) / 2.0
        eccentricity = math.sqrt(max(a*a - b*b, 0.0)) / (a + 1e-8) if a > 0 else 0.0
    else:
        eccentricity = 0.0

    moments = cv2.moments(mask)
    hu = cv2.HuMoments(moments).flatten()
    hu_log = [(-1 if h < 0 else 1) * math.log10(abs(h) + 1e-30) for h in hu]

    feats = {
        "area": float(area / (h * w)),
        "perimeter": float(perimeter / (h + w)),
        "form_factor": float(form_factor),
        "aspect_ratio": float(aspect_ratio),
        "extent": float(extent),
        "solidity": float(solidity),
        "eccentricity": float(eccentricity),
    }
    for i, v in enumerate(hu_log):
        feats[f"hu_{i+1}"] = float(v)
    return feats


def texture_features(gray: np.ndarray, mask: np.ndarray) -> Dict[str, float]:
    """GLCM (mean multi-jarak & sudut) + LBP uniform histogram."""
    gray_masked = np.where(mask > 0, gray, 0)
    gray_q = (gray_masked / 32).astype(np.uint8)  # quantize 0..7
    distances = [1, 2, 4]
    angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]

    glcm = graycomatrix(gray_q, distances=distances, angles=angles, levels=8, symmetric=True, normed=True)
    props = {}
    for prop in ["contrast", "dissimilarity", "homogeneity", "ASM", "energy", "correlation"]:
        vals = graycoprops(glcm, prop)
        props[prop] = float(np.mean(vals))

    P, R = 8, 1
    lbp = local_binary_pattern(gray_masked, P=P, R=R, method="uniform")
    lbp_masked = lbp[mask > 0]
    n_bins = P + 2
    hist, _ = np.histogram(lbp_masked, bins=np.arange(0, n_bins + 1), density=True)
    lbp_feats = {f"lbp_{i}": float(hist[i]) for i in range(n_bins)}

    out = {f"glcm_{k}": v for k, v in props.items()}
    out.update(lbp_feats)
    return out


def color_features(bgr: np.ndarray, mask: np.ndarray) -> Dict[str, float]:
    hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)
    mask3 = cv2.merge([mask, mask, mask])
    hsv_masked = cv2.bitwise_and(hsv, mask3)

    valid = mask > 0
    if np.count_nonzero(valid) == 0:
        return {"h_mean": 0.0, "s_mean": 0.0, "v_mean": 0.0, "h_std": 0.0, "s_std": 0.0, "v_std": 0.0}

    H = hsv[:, :, 0][valid]
    S = hsv[:, :, 1][valid]
    V = hsv[:, :, 2][valid]
    return {
        "h_mean": float(np.mean(H)), "s_mean": float(np.mean(S)), "v_mean": float(np.mean(V)),
        "h_std": float(np.std(H)),  "s_std": float(np.std(S)),  "v_std": float(np.std(V)),
    }


def extract_manual_features_from_bgr(bgr: np.ndarray) -> Dict[str, float]:
    mask = segment_leaf_mask(bgr)
    gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
    feats = {}
    feats.update(shape_features(mask))
    feats.update(texture_features(gray, mask))
    feats.update(color_features(bgr, mask))
    return feats


def extract_manual_features(image_path: str) -> Dict[str, float]:
    bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if bgr is None:
        raise ValueError(f"Failed to read image: {image_path}")
    feats = extract_manual_features_from_bgr(bgr)
    feats["filepath"] = image_path
    return feats


def build_features_dataframe(paths: List[str], labels: List[str]) -> pd.DataFrame:
    rows = []
    for p, y in tqdm(list(zip(paths, labels)), total=len(paths), desc="Extracting manual features"):
        try:
            feats = extract_manual_features(p)
            feats["label"] = y
            rows.append(feats)
        except Exception as e:
            print(f"[WARN] skip {p} -> {e}")
    df = pd.DataFrame(rows)
    return df
```

```python
import os, math
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Jika belum ada di kode kamu
MIN_AREA_RATIO = 0.005

# >>> Fungsi segment_leaf_mask() dan shape_features() DIASUMSIKAN sudah ada <<<

def visualize_shape_metrics_per_image(image_path: str):
    # --- Baca gambar ---
    bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if bgr is None:
        raise FileNotFoundError(f"Gagal membaca gambar: {image_path}")
    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    h, w = rgb.shape[:2]

    # --- Mask & kontur terbesar ---
    mask = segment_leaf_mask(bgr)
    cnts, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(cnts) == 0:
        raise RuntimeError("Tidak ditemukan kontur pada mask. Coba gambar lain / cek segmentasi.")
    areas = [cv2.contourArea(c) for c in cnts]
    c = cnts[int(np.argmax(areas))]

    # --- Hitung metrik mentah (px) + metrik turunan ---
    area_px = cv2.contourArea(c)
    perimeter_px = cv2.arcLength(c, True)
    x, y, bw, bh = cv2.boundingRect(c)
    aspect_ratio = bw / (bh + 1e-8)
    extent = area_px / (bw * bh + 1e-8)

    hull = cv2.convexHull(c)
    hull_area = cv2.contourArea(hull)
    solidity = area_px / (hull_area + 1e-8) if hull_area > 0 else 0.0

    if len(c) >= 5:
        (xc, yc), (MA, ma), ang = cv2.fitEllipse(c)
        a = max(MA, ma) / 2.0
        b = min(MA, ma) / 2.0
        eccentricity = math.sqrt(max(a*a - b*b, 0.0)) / (a + 1e-8) if a > 0 else 0.0
    else:
        xc, yc, MA, ma, ang = w/2, h/2, 0, 0, 0
        a = b = 0.0
        eccentricity = 0.0

    form_factor = 4 * math.pi * area_px / (perimeter_px**2 + 1e-8)

    # --- Versi normalisasi dari fungsi kamu (opsional untuk referensi) ---
    feats_norm = shape_features(mask)  # berisi area/perimeter ter-normalisasi dll

    # ===================== Visualisasi per metrik =====================

    # 0) Original
    fig, axes = plt.subplots(2, 4, figsize=(18, 9))
    ax = axes[0, 0]
    ax.imshow(rgb); ax.set_title("Gambar Asli"); ax.axis("off")

    # 1) Area (mask terisi)
    area_img = np.zeros_like(rgb)
    area_img[mask > 0] = (0, 255, 0)  # hijau untuk area daun
    blend_area = cv2.addWeighted(rgb, 0.6, area_img, 0.4, 0)
    ax = axes[0, 1]
    ax.imshow(blend_area)
    ax.set_title(f"Area (px): {int(area_px)}\nArea(norm): {feats_norm['area']:.4f}")
    ax.axis("off")

    # 2) Perimeter (kontur)
    perim_img = rgb.copy()
    cv2.drawContours(perim_img, [c], -1, (255, 0, 0), 3)  # biru
    ax = axes[0, 2]
    ax.imshow(perim_img)
    ax.set_title(f"Perimeter (px): {perimeter_px:.2f}\nPerim(norm): {feats_norm['perimeter']:.4f}")
    ax.axis("off")

    # 3) Form Factor (kontur + lingkaran setara luas)
    ff_img = rgb.copy()
    # centroid
    M = cv2.moments(c)
    if M["m00"] != 0:
        cx, cy = int(M["m10"] / M["m00"]), int(M["m01"] / M["m00"])
    else:
        cx, cy = w // 2, h // 2
    # radius lingkaran ekuivalen (luas sama)
    r_eq = int(np.sqrt(area_px / math.pi))
    cv2.drawContours(ff_img, [c], -1, (255, 0, 0), 2)  # kontur biru
    # lingkaran referensi (putih)
    cv2.circle(ff_img, (cx, cy), r_eq, (255, 255, 255), 2)
    ax = axes[0, 3]
    ax.imshow(ff_img)
    ax.set_title(f"Form Factor: {form_factor:.4f}\n(= 4πA / P²)")
    ax.axis("off")

    # 4) Aspect Ratio (bbox + anotasi W/H)
    ar_img = rgb.copy()
    cv2.rectangle(ar_img, (x, y), (x + bw, y + bh), (0, 255, 0), 2)  # bbox hijau
    # garis lebar (W)
    cv2.line(ar_img, (x, y + bh + 10), (x + bw, y + bh + 10), (0, 255, 0), 2)
    cv2.putText(ar_img, f"W={bw}", (x, y + bh + 30), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)
    # garis tinggi (H)
    cv2.line(ar_img, (x - 10, y), (x - 10, y + bh), (0, 255, 0), 2)
    cv2.putText(ar_img, f"H={bh}", (x - 10, y - 10 if y-10>10 else y + 15), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)
    ax = axes[1, 0]
    ax.imshow(ar_img)
    ax.set_title(f"Aspect Ratio (W/H): {aspect_ratio:.4f}")
    ax.axis("off")

    # 5) Extent (mask + bbox, persentase cakupan)
    ext_img = rgb.copy()
    # highlight area di dalam bbox
    overlay = ext_img.copy()
    roi = (mask > 0).astype(np.uint8)
    color_layer = np.zeros_like(ext_img)
    color_layer[roi > 0] = (0, 255, 0)  # hijau
    ext_highlight = cv2.addWeighted(ext_img, 0.6, color_layer, 0.4, 0)
    cv2.rectangle(ext_highlight, (x, y), (x + bw, y + bh), (255, 255, 255), 2)
    ax = axes[1, 1]
    ax.imshow(ext_highlight)
    ax.set_title(f"Extent: {extent:.4f}  (A / (W×H))")
    ax.axis("off")

    # 6) Solidity (kontur vs convex hull)
    sol_img = rgb.copy()
    overlay = sol_img.copy()
    # isi hull semi-transparan oranye
    cv2.drawContours(overlay, [hull], -1, (255, 165, 0), thickness=cv2.FILLED)  # oranye
    sol_img = cv2.addWeighted(overlay, 0.3, sol_img, 0.7, 0)
    # kontur daun (biru)
    cv2.drawContours(sol_img, [c], -1, (0, 0, 255), 2)
    ax = axes[1, 2]
    ax.imshow(sol_img)
    ax.set_title(f"Solidity: {solidity:.4f}  (A / A_hull)")
    ax.axis("off")

    # 7) Eccentricity (ellipse fit + sumbu utama/minor)
    ecc_img = rgb.copy()
    if len(c) >= 5 and a > 0 and b > 0:
        # ellipse
        cv2.ellipse(ecc_img, ((int(xc), int(yc)), (int(MA), int(ma)), ang), (255, 0, 255), 2)  # magenta
        # sumbu
        theta = np.deg2rad(ang)
        # mayor
        x1 = int(xc + a * np.cos(theta)); y1 = int(yc + a * np.sin(theta))
        x2 = int(xc - a * np.cos(theta)); y2 = int(yc - a * np.sin(theta))
        cv2.line(ecc_img, (x1, y1), (x2, y2), (255, 0, 255), 2)
        # minor
        x3 = int(xc + b * np.cos(theta + np.pi/2)); y3 = int(yc + b * np.sin(theta + np.pi/2))
        x4 = int(xc - b * np.cos(theta + np.pi/2)); y4 = int(yc - b * np.sin(theta + np.pi/2))
        cv2.line(ecc_img, (x3, y3), (x4, y4), (255, 0, 255), 2)
    ax = axes[1, 3]
    ax.imshow(ecc_img)
    ax.set_title(f"Eccentricity: {eccentricity:.4f}")
    ax.axis("off")

    plt.tight_layout()
    plt.show()

    # --- Tabel ringkas angka metrik ---
    df = pd.DataFrame([
        ["area_px", area_px],
        ["perimeter_px", perimeter_px],
        ["form_factor", form_factor],
        ["aspect_ratio", aspect_ratio],
        ["extent", extent],
        ["solidity", solidity],
        ["eccentricity", eccentricity],
        ["area_norm", feats_norm["area"]],
        ["perimeter_norm", feats_norm["perimeter"]],
    ], columns=["feature", "value"])
    print(df.to_string(index=False))


# ========= CONTOH PAKAI =========
TEST_IMG = r"sa5.jpg"  # ganti ke gambar kamu
visualize_shape_metrics_per_image(TEST_IMG)
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# >>> diasumsikan sudah ada:
# - segment_leaf_mask(bgr)
# - color_features(bgr, mask)

def visualize_color_features(image_path: str):
    # --- baca gambar ---
    bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if bgr is None:
        raise FileNotFoundError(f"Gagal membaca gambar: {image_path}")
    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)

    # --- mask daun ---
    mask = segment_leaf_mask(bgr)
    valid = mask > 0
    n_valid = int(np.count_nonzero(valid))
    if n_valid == 0:
        raise RuntimeError("Mask tidak memiliki piksel valid. Cek kembali segmentasi/ gambar.")

    # --- fitur warna (mean & std) ---
    feats = color_features(bgr, mask)
    h_mean, s_mean, v_mean = feats["h_mean"], feats["s_mean"], feats["v_mean"]
    h_std , s_std , v_std  = feats["h_std"] , feats["s_std"] , feats["v_std"]

    # --- siapkan peta H/S/V hanya pada area daun ---
    H = hsv[:, :, 0].astype(np.float32)
    S = hsv[:, :, 1].astype(np.float32)
    V = hsv[:, :, 2].astype(np.float32)

    # untuk visual: nolkan area non-daun
    H_vis = np.where(valid, H, np.nan)
    S_vis = np.where(valid, S, np.nan)
    V_vis = np.where(valid, V, np.nan)

    # --- buat swatch warna dari mean HSV ---
    swatch_hsv = np.zeros((50, 50, 3), dtype=np.uint8)
    swatch_hsv[..., 0] = np.clip(h_mean, 0, 179).astype(np.uint8)
    swatch_hsv[..., 1] = np.clip(s_mean, 0, 255).astype(np.uint8)
    swatch_hsv[..., 2] = np.clip(v_mean, 0, 255).astype(np.uint8)
    swatch_rgb = cv2.cvtColor(swatch_hsv, cv2.COLOR_HSV2RGB)

    # ================== plot grid ==================
    fig = plt.figure(figsize=(16, 10))

    # 1) gambar asli
    ax = plt.subplot(2, 3, 1)
    ax.imshow(rgb)
    ax.set_title("Gambar Asli")
    ax.axis("off")

    # 2) peta Hue
    ax = plt.subplot(2, 3, 2)
    im = ax.imshow(H_vis, cmap="hsv", vmin=0, vmax=179)
    ax.set_title("Hue (0–179) — area daun saja")
    ax.axis("off")
    cbar = plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    cbar.ax.set_ylabel("Hue")

    # 3) peta Saturation
    ax = plt.subplot(2, 3, 3)
    im = ax.imshow(S_vis, cmap="viridis", vmin=0, vmax=255)
    ax.set_title("Saturation (0–255)")
    ax.axis("off")
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    # 4) peta Value
    ax = plt.subplot(2, 3, 4)
    im = ax.imshow(V_vis, cmap="gray", vmin=0, vmax=255)
    ax.set_title("Value (0–255)")
    ax.axis("off")
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    # 5) histogram H/S/V
    ax = plt.subplot(2, 3, 5)
    Hv = H[valid].ravel()
    Sv = S[valid].ravel()
    Vv = V[valid].ravel()
    ax.hist(Hv, bins=32, range=(0, 179), alpha=0.7, label="Hue")
    ax.hist(Sv, bins=32, range=(0, 255), alpha=0.7, label="Saturation")
    ax.hist(Vv, bins=32, range=(0, 255), alpha=0.7, label="Value")
    ax.set_title("Histogram HSV (area daun)")
    ax.set_xlabel("Nilai")
    ax.set_ylabel("Frekuensi")
    ax.legend()

    # 6) swatch + ringkasan angka
    ax = plt.subplot(2, 3, 6)
    ax.imshow(swatch_rgb)
    ax.axis("off")
    ax.set_title(
        "Swatch (mean HSV → RGB)\n"
        f"H: {h_mean:.2f}±{h_std:.2f} | "
        f"S: {s_mean:.2f}±{s_std:.2f} | "
        f"V: {v_mean:.2f}±{v_std:.2f}\n"
        f"piksel valid: {n_valid}"
    )

    plt.tight_layout()
    plt.show()

    # --- cetak nilai fitur ringkas ---
    print("=== Fitur Warna (HSV) ===")
    print(f"H mean={h_mean:.4f}, std={h_std:.4f}")
    print(f"S mean={s_mean:.4f}, std={s_std:.4f}")
    print(f"V mean={v_mean:.4f}, std={v_std:.4f}")

# ========= CONTOH PAKAI =========
TEST_IMG = r"sa5.jpg"  # ganti ke file-mu
visualize_color_features(TEST_IMG)
```

```python
import os
import cv2
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Pastikan import skimage (dipakai di texture_features)
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern

# Jika belum ada di kode kamu (untuk segment_leaf_mask)
try:
    MIN_AREA_RATIO
except NameError:
    MIN_AREA_RATIO = 0.005

# >>> diasumsikan sudah ada:
# - segment_leaf_mask(bgr)
# - texture_features(gray, mask)
#   (implementasimu memakai quantisasi 8 level & distances=[1,2,4], angles=[0,45,90,135] deg)

def visualize_texture_features(image_path: str):
    # --- baca gambar & siapkan kanal ---
    bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if bgr is None:
        raise FileNotFoundError(f"Gagal membaca gambar: {image_path}")
    rgb  = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)

    # --- mask daun ---
    mask = segment_leaf_mask(bgr)
    valid = mask > 0
    if np.count_nonzero(valid) == 0:
        raise RuntimeError("Mask kosong. Cek segmentasi / ganti gambar.")

    # --- hitung fitur tekstur dengan fungsi kamu ---
    feats = texture_features(gray, mask)
    # pecah ringkas GLCM props & LBP hist
    glcm_keys = ["glcm_contrast","glcm_dissimilarity","glcm_homogeneity",
                 "glcm_ASM","glcm_energy","glcm_correlation"]
    glcm_vals = [feats[k] for k in glcm_keys]
    # siapkan LBP hist (P=8 -> n_bins=10 => label 0..9)
    lbp_bins = sorted([k for k in feats.keys() if k.startswith("lbp_")],
                      key=lambda x: int(x.split("_")[1]))
    lbp_vals = [feats[k] for k in lbp_bins]

    # --- bangun ulang material untuk visualisasi yang koheren dengan fungsi kamu ---
    # Quantisasi 8 level (0..7) seperti di fungsi
    gray_masked = np.where(valid, gray, 0)
    gray_q = (gray_masked / 32).astype(np.uint8)  # 0..7

    distances = [1, 2, 4]
    angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]
    glcm = graycomatrix(gray_q, distances=distances, angles=angles,
                        levels=8, symmetric=True, normed=True)  # shape: (8,8,len(d),len(a))

    # Heatmap GLCM per jarak -> rata-rata atas sudut (axis=-1)
    glcm_by_d = [np.nanmean(glcm[:, :, i, :], axis=-1) for i in range(len(distances))]  # 3 matriks 8x8

    # LBP uniform map (untuk visual)
    P, R = 8, 1
    lbp_map = local_binary_pattern(gray_masked, P=P, R=R, method="uniform")
    lbp_vis = np.where(valid, lbp_map, np.nan)  # non-daun disembunyikan

    # ================== Plot grid ==================
    fig = plt.figure(figsize=(18, 12))

    # 1) Gambar asli
    ax = plt.subplot(3, 3, 1)
    ax.imshow(rgb)
    ax.set_title("Gambar Asli")
    ax.axis("off")

    # 2) Gray (mask applied)
    ax = plt.subplot(3, 3, 2)
    gray_vis = gray.astype(np.float32)
    gray_vis[~valid] = np.nan
    im = ax.imshow(gray_vis, cmap="gray", vmin=0, vmax=255)
    ax.set_title("Grayscale (area daun saja)")
    ax.axis("off")
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    # 3) Peta LBP (uniform, label 0..P+1)
    ax = plt.subplot(3, 3, 3)
    im = ax.imshow(lbp_vis, cmap="viridis", vmin=0, vmax=P+1)
    ax.set_title("Peta LBP (uniform) — area daun")
    ax.axis("off")
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    # 4–6) Heatmap GLCM rata-rata sudut untuk d=1,2,4
    for idx, (d, mat) in enumerate(zip(distances, glcm_by_d), start=4):
        ax = plt.subplot(3, 3, idx)
        im = ax.imshow(mat, cmap="magma", vmin=0, vmax=np.max(mat) if np.max(mat)>0 else 1)
        ax.set_title(f"GLCM Heatmap (d={d}), avg 4 sudut\n(levels=8)")
        ax.set_xlabel("Gray level j")
        ax.set_ylabel("Gray level i")
        plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    # 7) Histogram LBP
    ax = plt.subplot(3, 3, 7)
    ax.bar(range(len(lbp_vals)), lbp_vals)
    ax.set_xticks(range(len(lbp_vals)))
    ax.set_xticklabels([b.replace("lbp_", "") for b in lbp_bins])
    ax.set_title("Histogram LBP (P=8, uniform)")
    ax.set_xlabel("Label LBP")
    ax.set_ylabel("Probabilitas")

    # 8) Ringkasan GLCM props
    ax = plt.subplot(3, 3, 8)
    ax.bar(range(len(glcm_keys)), glcm_vals)
    ax.set_xticks(range(len(glcm_keys)))
    ax.set_xticklabels([k.replace("glcm_", "") for k in glcm_keys], rotation=20)
    ax.set_title("Ringkasan Metrik GLCM (rata2 d & sudut)")
    ax.set_ylabel("Nilai")

    # 9) Mask (untuk referensi cepat)
    ax = plt.subplot(3, 3, 9)
    ax.imshow(mask, cmap="gray")
    ax.set_title("Mask Daun")
    ax.axis("off")

    plt.tight_layout()
    plt.show()

    # --- Tabel ringkasan numerik (rapi) ---
    df_glcm = pd.DataFrame({
        "metric": [k.replace("glcm_", "") for k in glcm_keys],
        "value":  [round(v, 6) for v in glcm_vals]
    })
    df_lbp = pd.DataFrame({
        "lbp_label": [int(b.split("_")[1]) for b in lbp_bins],
        "prob":      [round(v, 6) for v in lbp_vals]
    })
    print("=== GLCM (avg multi-jarak & sudut) ===")
    print(df_glcm.to_string(index=False))
    print("\n=== LBP Uniform Histogram ===")
    print(df_lbp.to_string(index=False))

# ========= CONTOH PAKAI =========
TEST_IMG = r"sa5.jpg"  # ganti ke file-mu
visualize_texture_features(TEST_IMG)
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def smart_interpolation(src_shape, target_size):
    """Pilih interpolasi: AREA untuk mengecilkan, CUBIC untuk membesarkan."""
    (h0, w0) = src_shape[:2]
    (tw, th) = target_size
    if tw < w0 or th < h0:
        return cv2.INTER_AREA
    return cv2.INTER_CUBIC

def compute_stats(img):
    """Hitung min/max/mean per channel (BGR atau RGB)."""
    if img.ndim == 2:
        img = img[..., None]
    stats = []
    for ch in range(img.shape[2]):
        c = img[..., ch]
        stats.append({
            "ch": ch,
            "min": float(np.nanmin(c)),
            "max": float(np.nanmax(c)),
            "mean": float(np.nanmean(c))
        })
    return pd.DataFrame(stats)

def visualize_resize_and_normalize(image_path: str, target_size=(224, 224)):
    # --- Baca gambar ---
    bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if bgr is None:
        raise FileNotFoundError(f"Gagal membaca gambar: {image_path}")
    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

    # --- Resize ---
    inter = smart_interpolation(rgb.shape, (target_size[0], target_size[1]))
    resized = cv2.resize(rgb, target_size, interpolation=inter)

    # --- Normalisasi (0..1) ---
    normalized = resized.astype(np.float32) / 255.0

    # --- Statistik per-channel ---
    stats_before = compute_stats(rgb)     # 0..255
    stats_resized = compute_stats(resized)
    stats_after  = compute_stats(normalized)  # 0..1

    # ================== VISUAL: Gambar ==================
    plt.figure(figsize=(15, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(rgb)
    plt.title(f"Asli\n{rgb.shape[1]}x{rgb.shape[0]}")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.imshow(resized)
    plt.title(f"Resize → {target_size[0]}×{target_size[1]}")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    # normalized di-range 0..1 (matplotlib oke langsung)
    plt.imshow(np.clip(normalized, 0, 1))
    plt.title("Normalisasi (0–1)")
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    # ================== VISUAL: Histogram ==================
    # Histogram sebelum normalisasi (RGB 0..255)
    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    for i, label in enumerate(["R", "G", "B"]):
        plt.hist(rgb[..., i].ravel(), bins=32, range=(0, 255), alpha=0.6, label=label)
    plt.title("Histogram RGB - Sebelum (0–255)")
    plt.xlabel("Nilai")
    plt.ylabel("Frekuensi")
    plt.legend()

    # Histogram sesudah normalisasi (0..1)
    plt.subplot(1, 2, 2)
    for i, label in enumerate(["R", "G", "B"]):
        plt.hist(normalized[..., i].ravel(), bins=32, range=(0.0, 1.0), alpha=0.6, label=label)
    plt.title("Histogram RGB - Sesudah (0–1)")
    plt.xlabel("Nilai (ter-normalisasi)")
    plt.ylabel("Frekuensi")
    plt.legend()

    plt.tight_layout()
    plt.show()

    # ================== CETAK STATISTIK ==================
    print("=== Statistik Sebelum (RGB 0–255) ===")
    print(stats_before.rename(columns={"ch":"channel(0=R,1=G,2=B)"}).to_string(index=False))

    print("\n=== Statistik Setelah Resize (RGB 0–255) ===")
    print(stats_resized.rename(columns={"ch":"channel(0=R,1=G,2=B)"}).to_string(index=False))

    print("\n=== Statistik Setelah Normalisasi (0–1) ===")
    print(stats_after.rename(columns={"ch":"channel(0=R,1=G,2=B)"}).to_string(index=False))

# ====== CONTOH PAKAI ======
TEST_IMG = r"sa5.jpg"   # ganti ke filemu
visualize_resize_and_normalize(TEST_IMG, target_size=(224, 224))
```
